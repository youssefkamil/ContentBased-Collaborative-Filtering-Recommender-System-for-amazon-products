**Steps:**
- Get products dataframe with categoties (given data)

- Weighing the categories (input user ratings * user products categories)
- Get input user profile (sum of user weighted categories)
- Find the recommendations (user profile * original products categories)

```python
products_df['categories'] = products_df['categories'].str.split(',')
products_df.head()
```

```python
products_df=products_df.drop_duplicates('name')
products_df=products_df[['id','name','categories']]
products_df
```

```python
productsWithCatg_df = products_df[['id','name','categories']].copy()
productsWithCatg_df.head()
```

```python
for index, row in products_df.iterrows():  
    for category in row['categories']:
        productsWithCatg_df.at[index, category] = 1

productsWithCatg_df.head()
```

```python
print("number of products : " , len(productsWithCatg_df['name'].unique()))
```

```python
productsWithCatg_df= productsWithCatg_df.drop_duplicates('name')
productsWithCatg_df
```

```python
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
productsWithCatg_df = productsWithCatg_df.fillna(0)
productsWithCatg_df.head()
```

## 2. Get input user ratings

We will implement __Content-Based recommendation systems__.

This technique attempts to figure out what user's favourite movie genres are, and then recommends movies that present those genres.

```python
userInput = [
            {'name':productsWithCatg_df['name'].unique()[0], 'rating':1},
            {'name':productsWithCatg_df['name'].unique()[7],'rating':3.5},
            {'name':productsWithCatg_df['name'].unique()[20],'rating':4},
            {'name':productsWithCatg_df['name'].unique()[24], 'rating':5},
            {'name':productsWithCatg_df['name'].unique()[28],'rating':2.5},
            ] 
inputProducts = pd.DataFrame(userInput)
inputProducts
```

#### Add movieId to input user
We will extract the input movies' ID's from the movies dataframe and add them into it.

```python
#Filtering out the movies by title
inputId = products_df[products_df['name'].isin(inputProducts['name'].tolist())]
inputId
```

```python
#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputProducts = pd.merge(inputId, inputProducts)
#Dropping information we won't use from the input dataframe
# inputMovies = inputMovies.drop(['genres', 'year'], axis=1)
#Final input dataframe
inputProducts
```

## 3. Weighing user categories
## 4. Get input user profile

Get the subset of movies that the input has watched from the Dataframe containing genres.

```python
#Filtering out the movies from the input
userProducts = productsWithCatg_df[productsWithCatg_df['name'].isin(inputProducts['name'].tolist())]
userProducts
```

We only need the actual genre table, so let's clean this up a bit by resetting the index and dropping the movieId, title, genres and year columns.

```python
#Resetting the index to avoid future issues
userProducts = userProducts.reset_index(drop=True)
#Dropping unnecessary columns
userCatTable = userProducts.drop(['id', 'name', 'categories'], axis=1)
userCatTable
```

We start learning the input's preferences.

```python
inputProducts['rating']
```

```python
userInput
```

```python
userCatTable
```

```python
#Dot product to get weighted genre matrix
userProfile = userCatTable.transpose().dot(inputProducts['rating'])
#The user profile
userProfile
```

## 5. Find the recommendations

Extracting the genre table from the original dataframe:

```python
productsWithCatg_df
```

```python
#Now let's get the genres of every movie in our original dataframe
CategoryTable = productsWithCatg_df.set_index(productsWithCatg_df['name'])
#And drop the unnecessary columns
CategoryTable = CategoryTable.drop(['id', 'name', 'categories'], axis=1)
CategoryTable
```

With the input's profile and the complete list of movies and their genres, we're going to take the weighted average of every movie based on the input profile and recommend the top twenty movies that most satisfy it.

```python
#test
((CategoryTable*userProfile).sum(axis=1))/(userProfile.sum())
```

```python
#Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((CategoryTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df
```

```python
recommendationTableWeighted_df = (recommendationTable_df.reset_index()).rename(columns={0: 'weightedAvg'})
recommendationTableWeighted_df
```

The final recommendation table

```python
user_recommendationTable_df = pd.merge(products_df, recommendationTableWeighted_df)
user_recommendationTable_df
```

The top 20 recommended movies!

```python
#Sort our recommendations in descending order
user_recommendationTable_df = user_recommendationTable_df.sort_values(by='weightedAvg', ascending=False)
user_recommendationTable_df.head(20)
```

### Advantages and Disadvantages of Content-Based Filtering

##### Advantages
* Learns user's preferences
* Highly personalized for the user

##### Disadvantages
* Doesn't take into account what others think of the item, so low quality item recommendations might happen
* Determining what characteristics of the item the user dislikes or likes is not always obvious
